```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox
import numpy as np

# Function to process responses and give an answer
def process_response():
    
    try:
        # Get age input and validate
        age = int(age_entry.get())
        if age < 18 or age > 70:
            messagebox.showerror("Invalid input", "This assessment is based on data from 18 to 70 year olds.")
            return

        #Heart Disease/Risk Assessment Function
        def loan_approval(person_age, person_cholestrol, person_bp, person_wac):
            user_input = pd.DataFrame({'person_age': [person_age],'person_cholestrol': [person_cholestrol], 'person_bp': [person_bp], 'person_wac':[person_wac]})
            prediction = np.random.rand()
            #prediction = loan_xgb.predict(user_input)
            if prediction > 0.5:
                return 'Heart Disease Attack or Risk'
            else:
                return 'No Heart Disease Attack or Risk'

        # Make a prediction and print result
        result = loan_approval(person_age, person_cholestrol, person_bp, person_wac)
        print(result)
        messagebox.showinfo("Result", result)

    except ValueError:
        # Handle non-numeric input
        messagebox.showerror("Invalid Input", "Please enter a valid numeric input.")
        
#Create the main application window
app = tk.Tk()
app.title("Loan Approval Application")

#Header
tk.Label(app, text="Heart Disease Risk Assessment").pack()

#Age Input
tk.Label(app, text="Enter your age:").pack()
age_entry = tk.Entry(app)
age_entry.pack()

# Question 1: Age group request
tk.Label(app, text="What is your age group?").pack()
question1_var = tk.StringVar()
tk.Radiobutton(app, text="21-30", variable=question1_var, value="21-30").pack()
tk.Radiobutton(app, text="31-40", variable=question1_var, value="31-40").pack()
tk.Radiobutton(app, text="41-50", variable=question1_var, value="41-50").pack()
tk.Radiobutton(app, text="51-60", variable=question1_var, value="51-60").pack()
tk.Radiobutton(app, text="61-70", variable=question1_var, value="61-70").pack()
tk.Radiobutton(app, text="71-80", variable=question1_var, value="71-80").pack()

# Question 2: Gender request
tk.Label(app, text="What is your biological gender?").pack()
question2_var = tk.StringVar()
tk.Radiobutton(app, text="M", variable=question2_var, value="male").pack()
tk.Radiobutton(app, text="F", variable=question2_var, value="female").pack()

# Question 3: Locality status
tk.Label(app, text="How would you describe the area you live in?").pack()
question3_var = tk.StringVar()  
tk.Radiobutton(app, text="Rural", variable=question3_var, value="rural").pack()
tk.Radiobutton(app, text="Urban", variable=question3_var, value="urban").pack()

# Question 4: Marital status
tk.Label(app, text="What is your marital status?").pack()
question4_var = tk.StringVar()  
tk.Radiobutton(app, text="Single", variable=question4_var, value='single').pack()
tk.Radiobutton(app, text="Married", variable=question4_var, value= 'married').pack()

# Question 5: Depression status
tk.Label(app, text="Have you been clinically diagnosed with depression?").pack()
question5_var = tk.StringVar()  
tk.Radiobutton(app, text="Yes", variable=question5_var, value='yes').pack()
tk.Radiobutton(app, text="No", variable=question5_var, value= 'no').pack()

# Store user input
person_age = [age_entry]
person_age_group = [question1_var.get()]
person_gender = [question2_var.get()]
person_locality_status = [question3_var.get()]
person_marital_status = [question4_var.get()]
person_depression_status = [question5_var.get()]

# Heart Disease/Risk Assesment Function
def loan_approval(person_age, person_age_group, person_gender, person_locality_status, person_marital_status, person_depression_status):
    user_input = pd.DataFrame({'person_age': [person_age],'person_age_group': [person_age_group], 'person_gender': [person_gender], 'person_locality_status': [person_locality_status], 'person_marital_status':[person_marital_status], 'person_depression_status': [person_depression_status]})
    prediction = np.random.rand()
    #prediction = loan_xgb.predict(user_input)
    if prediction > 0.5:
        return 'Loan Approved'
    else:
        return 'Loan Denied'
    
# Submit Button
tk.Button(app, text="Submit", command=process_response).pack()

# Run the application
app.mainloop()
```
